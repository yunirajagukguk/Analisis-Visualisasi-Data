```python
# Import the pandas library as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Read 'police.csv' into a DataFrame named ri
ri = pd.read_csv('haberman.csv')
```

```python
cancer_sur = pd.read_csv("haberman.csv")
cancer_sur.head()
```

```python
cancer_sur = pd.read_csv("haberman.csv",
header=None, names=["Age","operation_year",
"axil_nodes_det","surv_status"])
```

```python
sns.FacetGrid(cancer_sur,hue="surv_status",
height = 5).map(sns.distplot,
"Age").add_legend()
```

```python
sns.FacetGrid(cancer_sur,hue="surv_status",
height = 5).map(sns.distplot,
"operation_year").add_legend()
```

```python
counts, bin_edges = np.histogram(cancer_sur["Age"],
bins=10,density=True)
plt.xlabel("Age")
pdf = counts/(sum(counts))
print("pdf=", pdf);
print("bin_edges", bin_edges);
cdf = np.cumsum(pdf)
print("cdf=", cdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
sns.boxplot(x="surv_status",
y ="Age",data= cancer_sur)
plt.show()
```

```python
sns.boxplot(x="surv_status",
y ="operation_year",data= cancer_sur)
plt.show()
```

```python
sns.violinplot(
    x="surv_status",
    y="operation_year",
    data=cancer_sur,
    inner="quartile"  # tampilkan Q1, median, Q3
)
```

```python
sns.pairplot(cancer_sur,hue =
"surv_status", height= 3)
plt.show()
```
