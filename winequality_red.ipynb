Analisis Fitur

```python
import pandas as pd

# Memuat dataset
df = pd.read_csv("winequality-red.csv")

# ANALISIS ALCOHOL
median_alcohol = df['alcohol'].median()

df['alcohol_level'] = df['alcohol'].apply(
    lambda x: 'high' if x >= median_alcohol else 'low'
)

print(df.groupby('alcohol_level')['quality'].mean())

# ANALISIS pH
median_ph = df['pH'].median()

df['pH_level'] = df['pH'].apply(
    lambda x: 'high' if x >= median_ph else 'low'
)

print(df.groupby('pH_level')['quality'].mean())

# ANALISIS RESIDUAL SUGAR
median_sugar = df['residual sugar'].median()

df['sugar_level'] = df['residual sugar'].apply(
    lambda x: 'high' if x >= median_sugar else 'low'
)

print(df.groupby('sugar_level')['quality'].mean())

# ANALISIS CITRIC ACID
median_citric = df['citric acid'].median()

df['citric_level'] = df['citric acid'].apply(
    lambda x: 'high' if x >= median_citric else 'low'
)

print(df.groupby('citric_level')['quality'].mean())
```

```python
#Distribusi Kualitas Wine
import matplotlib.pyplot as plt

df['quality'].value_counts().sort_index().plot(kind='bar')

plt.xlabel("Quality")
plt.ylabel("Jumlah Data")
plt.title("Distribusi Kualitas Wine")

plt.show()
```

Korelasi antar Fitur

```python
import matplotlib.pyplot as plt

corr = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10,8))
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")
plt.show()
```

Visualisasi Alcohol vs Quality

```python
import seaborn as sns

sns.boxplot(x='quality', y='alcohol', data=df)

plt.title("Alcohol vs Quality")
plt.show()
```
